```python
# importar os pacotes necessários
import pandas as pd
import folium
import matplotlib.pyplot as plt
```

```python
df_estados = pd.read_csv('data/estados.csv')
df_municipios = pd.read_csv('data/casos_municipios.csv', encoding='latin_1')
```

```python
df_estados.head()
```

```python
df_municipios.head()
```

```python
df =  pd.merge(df_estados, df_municipios, on='codigo_uf')
```

```python
df.head()
```

```python
df.rename(columns={'nome_x': 'estado', 'nome_y':'municipio', 'Casos': 'casos'}, inplace=True)
```

```python
df.head()
```

```python
#Verificando o dataframe importado 
fig, ax = plt.subplots(figsize=(10,10))

df.plot.scatter(x='longitude', y='latitude', ax=ax)
plt.tight_layout()
```

```python
df_rs = df[ (df['uf'] == 'RS') &  df['casos'].notna() ]
```

```python
df_rs.head(100)
```

```python
df_rs.describe()
```

```python
# extraindo os valore do dataframe para variáveis
latitudes = df_rs.latitude.values
longitudes = df_rs.longitude.values
municipios = df_rs.municipio.values
```

```python
# criando um mapa
mapa = folium.Map(location=[-29.772524, -53.240623],
                  zoom_start=6.45
                 )

#adicionado os marcadores no mapa
for lat, lon, mun in zip(df_rs.latitude.values, df_rs.longitude.values, df_rs.municipio.values):
  folium.Marker(location=[float(lat), float(lon)], tooltip=mun).add_to(mapa)
mapa
```

```python
# criando um mapa
mapa = folium.Map(location=[-29.772524, -53.240623],
                  zoom_start=6.45,  
                 )

#adicionado os marcadores no mapa
for lat, lon, mun, cas in zip(df_rs.latitude.values, df_rs.longitude.values, df_rs.municipio.values, df_rs.casos):
  folium.CircleMarker(
    location=[float(lat), float(lon)],
    radius=cas*1.5,
    popup= mun + ': ' + str(int(cas)) + ' casos',
    color='#FE3939',
    fill=True,
    fill_color='#FE8484'
  ).add_to(mapa)

mapa
```

```python
import seaborn as sns
sns.set()
```

```python
casos_rs = df_rs[['casos', 'municipio']].sort_values(by=['casos'], ascending=False)
casos_rs
```

```python
fig, ax = plt.subplots(figsize=(10,7))
sns.set_color_codes("pastel")
sns.barplot(x="casos", y="municipio", data=casos_rs, ax=ax, color="r")
ax.set_title("Número de Casos no RS")
ax.set_xlabel("Casos")
ax.set_ylabel("Município")

plt.tight_layout()
```
